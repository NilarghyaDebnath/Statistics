```python
from scipy.stats import binom

# Calculate probability of a customer accepting the offer
p = df['Response'].mean()  # Probability of success (accepting the offer)
n = len(df['Response'])    # Number of trials (total number of customers)
probability = binom.pmf(k=1, n=n, p=p)  # Probability of exactly 1 success (acceptance)
probability
```

```python
probability = binom.pmf(k=0, n=n, p=p)  # Probability of exactly 1 success (acceptance)
probability
```

```python
import datetime

# Get the current year
current_year = datetime.datetime.now().year

# Calculate age from year of birth
df['Age'] = current_year - df['Year_Birth']


from scipy.stats import norm

# Calculate probability of a customer having a certain age
mean_age = df['Age'].mean()
std_age = df['Age'].std()
probability_age_30 = norm.cdf(30, loc=mean_age, scale=std_age)  # Probability of age <= 30

print(probability_age_30)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select relevant features for clustering
features = ['Income','Recency','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth']
new_df = df[features]
new_df.head()
```

```python
# Standardize numerical features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(new_df)
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.metrics import silhouette_score

def silhouette_analysis(n,features):
  for n_clusters in range(2,n):
    km = KMeans(n_clusters=n_clusters,init="k-means++",max_iter=300,n_init=10,random_state=0)
    preds = km.fit_predict(features)
    centers = km.cluster_centers_

    score = silhouette_score(features, preds, metric='euclidean')
    print('For n_clusters = {}, silhouette score is {}'.format(n_clusters, score))

    visualizer = SilhouetteVisualizer(km)

    visualizer.fit(features)
    visualizer.poof()
```

```python
silhouette_analysis(11,df_scaled)
```

```python
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)
print(wcss)
```

```python
# Visualising elbow method
plt.plot(range(1,11),wcss, marker='o')
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.grid(True)
plt.show()
```

```python
# creating object kmeans of class KMeans()
kmeans=KMeans(n_clusters=4,init="k-means++",max_iter=300,n_init=10,random_state=0)
```

```python
new_df['Cluster'] = kmeans.fit_predict(df_scaled)

# it tells which data point belongs to which cluster (0,1,2,3)
```

```python
# Analyze characteristics of each cluster
cluster_summary = df.groupby('Cluster')[features].mean()

# Visualize cluster characteristics
cluster_summary.plot(kind='bar', figsize=(10, 6))
plt.title('Average Spending Habits by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Average Spending')
plt.xticks(rotation=0)
plt.legend(title='Product Category')
plt.show()
```

```python
# Convert new_df and cluster labels to numpy arrays
x1 = np.array(new_df.drop('Cluster', axis=1))  # Drop the 'Cluster' column
y1_kmeans = new_df['Cluster'].values

# Plotting the clusters
fig = plt.figure(figsize=(10, 6))
plt.scatter(x1[y1_kmeans == 0, 0], x1[y1_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(x1[y1_kmeans == 1, 0], x1[y1_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(x1[y1_kmeans == 2, 0], x1[y1_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(x1[y1_kmeans == 3, 0], x1[y1_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')

plt.title('Clusters of customers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
